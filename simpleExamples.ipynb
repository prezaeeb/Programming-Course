# Word2vec Pretrained networks

```python
import gensim.downloader as api

# Load pre-trained Word2Vec model (Google News)
model = api.load("word2vec-google-news-300")

# Get the vector for a word
word_vector = model["girl"]

# Find similar words
similar_words = model.most_similar("girl")

print("Vector for 'girl':", word_vector[:5])  # Print first 5 values for brevity
print("Words similar to 'girl':", similar_words)
```

# GloVe pretrained networks

```python
# Load pre-trained GloVe model (Common Crawl, 840B tokens, 300D)
model = api.load("glove-wiki-gigaword-300")

# Get the vector for a word
word_vector = model["king"]

# Find similar words
similar_words = model.most_similar("king")

print("Vector for 'king':", word_vector[:5])  # Print first 5 values
print("Words similar to 'king':", similar_words)
```

# Examples of training Word2Vec on a sample text corpus

```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

# Sample corpus (list of sentences)
corpus = [
    "King and queen rule the kingdom",
    "The prince and princess are royalty",
    "A queen is married to a king",
    "The kingdom has many knights"
]

# Tokenize sentences
tokenized_corpus = [simple_preprocess(sentence) for sentence in corpus]

# Train Word2Vec model
model = Word2Vec(sentences=tokenized_corpus, size=100, window=3, min_count=1, workers=4)
# newer gensim ---> vector_size=100

# Test model: find words similar to "king"
print(model.wv.most_similar("king"))

# Save and load model
model.save("word2vec.model")
loaded_model = Word2Vec.load("word2vec.model")
```
